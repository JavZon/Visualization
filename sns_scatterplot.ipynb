```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Creating a DataFrame
data = {
    'City': ['California', 'Berlin', 'Copenhagen'],
    'Jan': [10, 15, 20],
    'Feb': [12, 18, 22],
    'Mar': [14, 20, 24]
}
df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(10, 4))
sns.heatmap(df.set_index('City'), vmin=0, vmax=30, cmap='coolwarm', annot=True, fmt=".1f")
plt.title('Temperature in Cities')
plt.xlabel('Month')
plt.ylabel('City')
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.heatmap(df.set_index('City'), cmap='coolwarm', annot=True, fmt=".1f")
plt.title('Temperature in Cities')
plt.xlabel('Month')
plt.ylabel('City')
plt.show()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='year', y='price')
plt.show()
```

```python
df=df[df['year']!=1970]
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=df, x='year', y='price')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='year', y='price', hue='fuelType')
plt.show()
```
