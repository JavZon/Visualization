```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
np.random.seed(0)
scores = np.random.normal(loc=70, scale=10, size=100)  # Normally distributed scores
df=pd.DataFrame({'Scores':scores})
df2=df[(df['Scores']>70.1) & (df['Scores']<77.1)]
df2['Scores']=df2['Scores'].copy().astype(int)
df2.head()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data=df2, x='Scores', binwidth=0.58, kde=True, color='g')
plt.show()
```

```python
sns.displot(data=df2, x='Scores',height=3,aspect=2 ,binwidth=0.58, color='g',kde=True)
plt.show()
```

```python
sns.displot(data=df2, x='Scores',height=3,aspect=2 , color='g',kind='ecdf')
plt.show()
```

```python
names = ['John', 'Alice', 'Bob', 'Matt']
np.random.shuffle(names)

subjects = ['Math', 'English', 'Biology']
np.random.shuffle(subjects)

scores = np.random.randint(55, 101, size=40)

df = pd.DataFrame({
    'Names': np.repeat(names, 10)[:40],  # repeat each name 10 times to match 40 rows
    'Scores': scores,
    'Subject': np.tile(subjects, 14)[:40]  # repeat subjects to match 40 rows
})

df2=df.sample(25)
df.shape
```

```python
df2.Scores.value_counts()
```

```python
sns.displot(data=df2, x='Scores', height=3.5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df2, x='Scores',hue='Names', height=3.5, aspect=2, multiple='stack')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df2, x='Scores',col='Names', height=3, aspect=1, color='g')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df2, x='Scores', height=3.5, aspect=2, kind='kde')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df2, x='Scores', height=3.5, aspect=2, kind='kde', hue='Names', multiple='stack')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.kdeplot(data=df2, x='Scores')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.kdeplot(data=df2, x='Scores', hue='Names', multiple='stack')
plt.xticks(rotation=90)
plt.show()
```
