```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.figure(figsize=(8,4))
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])

plt.grid(True, axis='y')
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.grid(which='minor')
plt.minorticks_on()
plt.show()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df=df[df.year>=2010] # 2010-yildan keyin chiqqan mashinalarni saqlab qolamiz
df=df[df.transmission!='Other'] # Faqat avtomat, mexanik va yarim-avtomat mashinalarni olib qolamiz
df=df[df.fuelType!='Other'] # Faqat benzin, dizel va gibrid mashinalarni olib qolamiz
df.head()
```

```python
plt.figure(figsize=(12,6))

plt.subplot(2,1,1)
sns.histplot(data=df, x='price')
plt.title('Histogram plot')
plt.xticks(rotation=90)

plt.subplot(2,1,2)
sns.scatterplot(data=df, x='price',y='year',hue='fuelType')
plt.title('Price Vs Year comparision')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
sns.histplot(data=df, x='price')
plt.title('Histogram plot')
plt.xticks(rotation=90)

plt.subplot(2,2,2)
sns.scatterplot(data=df, x='price',y='year',hue='fuelType')
plt.title('Price Vs Year comparision')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.subplot(3, 2, 1)
plt.plot([1, 2, 3], [1, 2, 3], 'r-')
plt.title('Subplot 1')

plt.subplot(3, 2, 2)
plt.plot([1, 2, 3], [3, 2, 1], 'g-')
plt.title('Subplot 2')

plt.subplot(3, 2, 3)
plt.plot([1, 2, 3], [2, 3, 2], 'b-')
plt.title('Subplot 3')

plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 5, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2, 2, 1)
plt.plot(x, y1, 'r-')
plt.title('Subplot 1: Sine Wave')

plt.subplot(2, 2, 2)
plt.plot(x, y2, 'g-')
plt.title('Subplot 2: Cosine Wave')

plt.tight_layout()  # Adjust subplot parameters to give specified padding
plt.show()
```

```python
sns.scatterplot(data=df, x='price',y='mileage' ,color='b')
plt.xticks(rotation=90)
plt.show()
```

```python
fig,ax=plt.subplots(2,1, figsize=(8,8))
fig.suptitle('Grafiklar jamlanmasi')
sns.scatterplot(ax=ax[0],data=df, x='price',y='mileage' ,color='b')
plt.xticks(rotation=90)
sns.histplot(ax=ax[1],data=df, x='price', color='g',kde=True)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
fig,ax=plt.subplots(2,2, figsize=(12,6))
fig.suptitle('Grafik N1')
sns.scatterplot(ax=ax[0,0],data=df, x='price',y='mileage' ,color='b')
plt.xticks(rotation=90)
sns.histplot(ax=ax[0,1],data=df, x='price', color='g',kde=True)
plt.xticks(rotation=90)
sns.scatterplot(ax=ax[1,0], data=df, x='mileage', y='year', color='r')
plt.xticks(rotation=90)
sns.histplot(ax=ax[1,1],data=df, x='mileage', color='g')
plt.xticks(rotation=90)
plt.tight_layout()
```
