```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title('Example Plot')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16],color='m',linestyle='-.') # 1- grafikning rangi va stili
plt.plot([9,7,4,0.8], [4, 3, 2,1], color='y', linestyle='--')  # 2- grafikning rangi va stili
plt.xlabel('X axis') # x o'qining nomi
plt.ylabel('Y axis') # y o'qining nomi
plt.grid(color='b', linestyle='--') # setkaning rangi va stilini o'zgartiradi
plt.legend(['a','b'], loc='best',labelcolor='linecolor')
plt.show()
```

```python
plt.figure(figsize=(8,6))
line1=plt.plot([1, 2, 3, 4], [1, 4, 9, 16],color='m',linestyle='-.')
line2=plt.plot([9,7,4,0.8], [4, 3, 2,1], color='y', linestyle='--')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid()
plt.legend(['A data','B data'],loc='best',labelcolor='linecolor', title='Matn', facecolor="inherit",shadow=True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot([1, 2, 3, 4], [1, 4, 9, 16],color='m',linestyle='-.', label='a')
plt.plot([9,7,4,0.8], [4, 3, 2,1], color='y', linestyle='--', label='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid()
plt.legend()
plt.show()
```
