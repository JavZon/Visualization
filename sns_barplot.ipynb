```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
x = [2.5,3.5,5.5,7.5]
y = [10, 20, 25, 30]

plt.figure(figsize=(6,3))
plt.plot(x, y)
plt.xticks(ticks=[1,2,3,4])
plt.show()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/uzbekistan.csv", index_col=0)
df.columns = ['maydon','aholi']
df
```

```python
df.index
```

```python
plt.figure(figsize=(9,4.5))
colors = ['blue','green','red','purple','orange','yellow','brown','pink','cyan','magenta','teal','lavender','olive']

sns.barplot(x=df.index[:4], y=df['aholi'][:4])
plt.title('Uzb viloyatlaridagi aholi soni')
plt.xlabel('Viloyat')
plt.ylabel('Aholi soni (million kishi)')
plt.grid()
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(14,6))
colors = ['blue','green','red','purple','orange','yellow','brown','pink','cyan','magenta','teal','lavender','olive']

sns.barplot(x=df.index, y=df['aholi'],hue=df.index ,palette=colors)
plt.title('Uzb viloyatlaridagi aholi soni')
plt.xlabel('Viloyat')
plt.ylabel('Aholi soni (million kishi)')
plt.grid()
plt.xticks(rotation=90)
plt.show()

```

```python
plt.figure(figsize=(9,4.5))
colors = ['blue','green','red','purple']

sns.barplot(x=df.index[:4], y=df['aholi'][:4], hue=df.index[:4], palette=colors)
plt.title('Uzb viloyatlaridagi aholi soni')
plt.xlabel('Viloyat')
plt.ylabel('Aholi soni (million kishi)')
plt.grid()
plt.xticks(rotation=90)
plt.show()
```

```python
data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Math': [85, 70, 90, 65, 75],
    'Science': [80, 75, 85, 60, 70],
    'English': [90, 65, 75, 80, 85]
}

# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(6,3))
sns.barplot(x='Student', y='Math', data=df, hue='Student',palette=['b','k','r','g','y'])
plt.title('Math Scores of Students')
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.barplot(x='Student', y='Math', hue='English', data=df,legend=False, palette=['b','k','r','g','y'])
plt.title('Math Scores of Students with Science Scores Hue')
plt.show()
```
