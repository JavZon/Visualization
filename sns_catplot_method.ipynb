```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
data = {
    "City": ["Tashkent",'London','Paris',"Tashkent",'London','Paris',"Tashkent",'London','Paris'],
    "Month": ['Jan','Jan','Jan','Feb','Feb','Feb','Mar','Mar','Mar'],
    "Temp": [20,18,22,21,23,24,25,19,17]
}

df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(12,6))
sns.catplot(x="City", y="Temp", data=df,hue='Month' ,kind="point")
plt.title('Average temp in cities on Jan,Feb and March')
plt.xlabel("Cities")
plt.ylabel("Temp")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.catplot(x="City", y="Temp", data=df,hue='Month' ,kind="bar")
plt.title('Average temp in cities on Jan,Feb and March')
plt.xlabel("Cities")
plt.ylabel("Temp")
plt.show()
```
