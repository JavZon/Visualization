```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/flight_delays.csv")
df.head()
df['Month']=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
df.set_index('Month', inplace=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(12,4))
sns.barplot(x=df.index, y=df['AA'])
plt.title("American Airways' average delay minutes")
plt.xlabel('Months')
plt.ylabel('Minutes')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.heatmap(df,cmap='YlGnBu',linewidths=0.555, linecolor='k', annot=True)
plt.title("Average delay minutes for American flight companies")
plt.xlabel('Airlines')
plt.show()
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 70, 90],
    'Science': [80, 75, 85],
    'English': [90, 65, 75]
}

df = pd.DataFrame(data)
df
```

```python
df2=pd.melt(df, id_vars='Name',value_vars=['Math','English','Science'], var_name='Subject',value_name='Score')
df2
```

```python
sns.catplot(data=df2, kind='bar', x="Name", y="Score", hue="Subject")
plt.show()
```

```python
sns.catplot(data=df2, kind="bar", x="Score", y="Name", hue="Subject")
plt.show()
```

```python
data={'City':['Tashkent','Samarkand','Bukhara'],
      'January':[20,18,22],
      'February':[21,23,24],
      'March':[23,20,22]}
df=pd.DataFrame(data)
df
```

```python
df3=pd.melt(df, id_vars='City',value_vars=['January','February','March'],var_name='Month', value_name='Temperature')
df3
```

```python
sns.catplot(data=df3, kind='bar', x="City", y="Temperature", hue="Month")
plt.show()
```
