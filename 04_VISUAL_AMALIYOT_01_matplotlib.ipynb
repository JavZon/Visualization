## 4-VAZIFA. `df`dagi barcha `-` belgilarni `NaN` qiymat bilan almashtiring.

```python
# JAVOBNI SHUYERGA YOZING
df.replace({'-':np.nan}, inplace=True)   #1-usul
df.apply(lambda x : np.nan if x is '-' else x) #2-usul
df
```

```python
#@title To'g'ri javob
df = df.replace("-", np.nan)
```

## 5-VAZIFA. `df` ustunlari qiymatlarini matndan o'nlik songa (`float`) o'zgartiring.

```python
# JAVOBNI SHUYERGA YOZING
type(df.loc[2006,'T'])
df=df.astype('float')
type(df.loc[2006,'T'])
df.head()
# df.isnull().sum()
```

```python
#@title To'g'ri javob:
df = df.astype(dtype='float')
df.head(10)
```

##7-VAZIFA. Chiziqli grafiklar

### 7.1. `df`dan yillik o'rtacha haroratni ko'rsatuvchi ustunni `x` o'zgaruvchiga saqlang va `plt.plot()` yordamida chiziqli grafik chizing.

```python
# JAVOBNI SHUYERGA YOZING
import matplotlib.pyplot as plt
%matplotlib inline
x=df['T']
plt.plot(x)
plt.show()
```

```python
#@title To'g'ri javob
import matplotlib.pyplot as plt
%matplotlib inline
x=df['T']
plt.plot(x)
plt.show()
```

### 7.2. Yuqoridagi grafikni hajmini kattalashtiring, chiziq qalingligi, turi, rangi, markerlar turini istalgancha o'zgartiring

```python
# JAVOBNI SHUYERGA YOZING
x=df['T']
plt.figure(figsize=(16,8))
plt.plot(x, linestyle='--', linewidth=2.5, marker='x',
         color='m', alpha=0.8)
plt.show()
```

```python
#@title Javobga misol:
x=df['T']
plt.figure(figsize=(16,8))
plt.plot(x, linewidth=2, linestyle="--", marker='h', markerfacecolor='r')
plt.show()
```

### 7.3. Yuqoridagi grafikda x va y o'qlarini nomlang, grafikka ham umumiy nom bering

```python
# JAVOBNI SHUYERGA YOZING
x=df['T']
plt.figure(figsize=(16,8))
plt.plot(x, linestyle='--', linewidth=2.5, marker='x',
         color='m', alpha=0.8)
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.title('Average annual temperature in Tashkent')
plt.show()
```

```python
#@title To'g'ri javob
x=df['T']
plt.figure(figsize=(16,8))
plt.plot(x, linewidth=2, linestyle="--", marker='h', markerfacecolor='r')
plt.title("Toshkent shahri uchun yillik o'rtacha harorat")
plt.xlabel("Yil")
plt.ylabel("Harorat")
plt.show()
```

## 8-VAZIFA. Toshkent shahri uchun, har bir yilda o'rtacha min va max haroratlarni ko'rsatuvchi grafik chizing. **Grafikni faylga saqlab oling**

```python
# JAVOBNI SHUYERGA YOZING
x=df['TM']
y=df['Tm']
plt.figure(figsize=(14,7))
plt.plot(x, linestyle='--', color='g',marker='x', label='Average Max Temperature')
plt.plot(y, linestyle='-.',color='r',marker='o', label='Average Max Temperature')
plt.title('Average annaul max and min temperature in Tashkent')
plt.legend(loc=0, shadow=True, title='Average Temperature')
plt.grid()

plt.xlabel('Year')
plt.ylabel('Temperature')
plt.savefig('max_min_temp_tash.png')
plt.show()
```

```python
#@title To'g'ri javob
minT = df['Tm']
maxT = df['TM']
plt.figure(figsize=(16,8))
plt.plot(minT, linewidth=2, linestyle="--", marker='h', markerfacecolor='g', label="O'rtacha minimum harorat")
plt.plot(maxT, linewidth=2, linestyle="-", marker='v', markerfacecolor='r', label="O'rtacha maksimum harorat")
plt.grid()
plt.legend()
plt.title("Toshkent shahri uchun yillik o'rtacha min va max harorat")
plt.xlabel("Yil")
plt.ylabel("Harorat")
plt.savefig('tempMinMax.png')
plt.show()
```
