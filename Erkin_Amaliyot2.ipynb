```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_excel('/content/uybor.xlsx')
df.head(3)
```

```python
df.shape
df=df[df['rooms']<6]
df=df[df['size']<250]
df=df[df['price']<200000]
df.shape
```

```python
df_mean_price=df.groupby('district')[['price']].mean()
df_mean_price
df_mean_rooms=df.groupby('district')[['rooms']].mean()
df_mean_rooms
df_mean_size=df.groupby('district')[['size']].mean()
df_mean_size
```

```python
df_mean_size=df.groupby('district')[['size']].mean()
df_mean_size
```

```python
fig,ax=plt.subplots(2,2, figsize=(10,10))
fig.suptitle('Grafiklar jamlanmasi')

sns.barplot(ax=ax[0,0], data=df_mean_price, x=df_mean_price.index, y=df_mean_price['price'])
ax[0,0].set_title("Uylarning tumanlar bo'yicha o'rtacha narxlari")
ax[0,0].set_xticklabels(ax[0,0].get_xticklabels(), rotation=90)

sns.barplot(ax=ax[0,1], data=df_mean_rooms, x=df_mean_rooms.index, y=df_mean_rooms['rooms'])
ax[0,1].set_title("Uylarning tumanlar bo'yicha xonalar soni")
ax[0,1].set_xticklabels(ax[0,1].get_xticklabels(), rotation=90)

sns.barplot(ax=ax[1,0], data=df_mean_size, x=df_mean_size.index, y=df_mean_size['size'])
ax[1,0].set_title("Uylarning tumanlar bo'yicha o'rtacha hajmi (m.kv)")
ax[1,0].set_xticklabels(ax[0,1].get_xticklabels(), rotation=90)

sns.barplot(ax=ax[1,1], data=df, x='rooms', y='price')
ax[1,1].set_title("Toshkent shahridagi uylarning narxlari va xonalari nisbati")

plt.tight_layout()
plt.show()
```
