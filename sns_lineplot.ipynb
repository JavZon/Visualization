```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Create a DataFrame with complex data
np.random.seed(0)
index = pd.date_range('2024-01-01', periods=100)
data = {
    'A': np.random.normal(0, 1, 100),
    'B': np.random.normal(1, 2, 100),
    'C': np.random.normal(-1, 0.5, 100)
}
df = pd.DataFrame(data, index=index)

# Plotting with plt.plot()
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['A'], label='A')
plt.plot(df.index, df['B'], label='B')
plt.plot(df.index, df['C'], label='C')
plt.title('Line Plot with plt.plot()')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Plotting with seaborn.lineplot()
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, dashes=False)
plt.title('Line Plot with seaborn.lineplot()')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend(title='Variables', labels=['A', 'B', 'C'], shadow=True, loc=0)
plt.grid(True)
plt.show()
```
