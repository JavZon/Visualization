```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#JAVOBINGIZNI SHU YERGA YOZING
df=pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/tashtemp.csv', index_col=0)
df.head()
```

### 2-VAZIFA. `df` barcha ustunlarini chiziqli grafik ko'rinishida chiqaring

```python
#JAVOBINGIZNI SHU YERGA YOZING
plt.figure(figsize=(15,6))
sns.lineplot(data=df)
plt.xlabel('Yil')
plt.title("Toshkentning ob-havo ko'rsatkichlari")
plt.show()
```

### 3-VAZIFA. Grafiklar nomini to'g'rilang (`T`, `Tm` va hokazo matnlarni o'zbek tilida chiqaring.)

```python
#JAVOBINGIZNI SHU YERGA YOZING
cols=["Yillik o'rtacha harorat","Yillik o'rtacha maksimal harorat", "Yillik o'rtacha minimal harorat","Yillik o'rtacha shamol tezligi",
      "Yomg'irli kunlar soni","Qorli kunlar soni","Bo'ronli kunlar soni","Tumanli kunlar soni","To'fonli kunlar soni","Do'l yoqqan kunlar"]
df.columns=cols
df.head()
plt.figure(figsize=(15,6))
sns.lineplot(data=df)
plt.xlabel('Yil')
plt.title("Toshkentning ob-havo ko'rsatkichlari")
plt.legend(loc=0)
plt.show()
```

```python
#@title Kutilgan natija:
labels=["Yillik o'rtacha harorat",
        "Yillik o'rtacha maksimal harorat",
        "Yillik o'rtacha minimal harorat",
        "Yillik o'rtacha shamol tezligi",
        "Yomg'irli kunlar soni",
        "Bo'ronli kunlar soni",
        "Tumanli kunlar soni",
        "To'fonli kunlar soni",
        "Do'l yoqqan kunlar"
        ]

plt.figure(figsize=(16,6))
sns.lineplot(data=df, linewidth=2.0)
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
plt.xlabel("Yil")
plt.ylabel("Harorat")
plt.legend(labels, loc=1)
plt.show()
```

### 5-VAZIFA. Yillik o'rtacha shamol tezligini ustunli grafik ko'rinishida chiqaring.

```python
#JAVOBINGIZNI SHU YERGA YOZING
df["Yillik o'rtacha shamol tezligi"]
plt.figure(figsize=(10,3))
sns.barplot(x=df.index, y=df["Yillik o'rtacha shamol tezligi"], hue=df.index, legend=False, palette='dark:g')
plt.xlabel('Yil')
plt.xticks(rotation=90)
plt.ylabel("O'rtacha shamol tezligi(km/soat)")
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
plt.show()
```

```python
#@title Kutilgan natija:
plt.figure(figsize=(16,6))
sns.barplot(x=df.index, y=df['V'])
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
plt.ylabel("O'rtacha shamol tezligi (km/soat)")
plt.xlabel("Yillar")
plt.xticks(rotation=90)
plt.show()
```

### 6-VAZIFA. Yomg'rili, qorli, bo'ronli, tumanli va do'l yoqqan kunlar sonini heatmap ko'rinishida chiqaring.

```python
df2=df[["Yomg'irli kunlar soni", 'Qorli kunlar soni', "Bo'ronli kunlar soni",
       'Tumanli kunlar soni', "To'fonli kunlar soni", "Do'l yoqqan kunlar"]]
df2=df2.T
df2
```

```python
#JAVOBINGIZNI SHU YERGA YOZING
plt.figure(figsize=(30,5))
sns.heatmap(data=df2 ,cmap='YlGnBu', linewidth=0.5, linecolor='k', square=True)
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
yticks_labels = ["Yomg'ir","Qor","Bo'ron","Tuman","Do'l"]
plt.yticks(np.arange(5)+0.5, labels=yticks_labels)
plt.xticks(rotation=90)
#plt.yticks(rotation=0)
plt.show()
```

```python
#@title Kutilgan natija:
plt.figure(figsize=(30,5))
sns.heatmap(data=df[['RA','SN','TS','FG','GR']].T, cmap="YlGnBu", linewidths=1, square=True)
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
yticks_labels = ["Yomg'ir","Qor","Bo'ron","Tuman","Do'l"]
plt.yticks(np.arange(5)+0.5, labels=yticks_labels)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```
