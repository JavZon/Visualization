```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female'],
    'Subject': ['Math', 'Math', 'Science', 'Science', 'English'],
    'Score': [85, 70, 80, 75, 90]
}
df = pd.DataFrame(data)
df
# Create a nested bar plot using facets (subplots)
#sns.catplot(x='Student', y='Score', hue='Gender', col='Subject', kind='bar', data=df)
#plt.show()
```

```python
sns.catplot(x='Student', y='Score', hue='Subject', col='Gender', kind='bar', data=df)
plt.show()
```

```python
data = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
plt.figure(figsize=(7,3.5))
sns.heatmap(data, annot=True, fmt='.2f', cmap='inferno', linewidths=.5)
plt.title('Customized Heatmap')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.show()
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv",index_col='Date', usecols=['Date','Avila Adobe'], parse_dates=['Date'])
df.head()
```

```python
type(df.index)
```

```python
print(df.index[2].year)
print(df.index[2].month)
print(df.index[2].day)
print(df.index[2].weekday())
print(df.index[2].hour)
print(df.index[2].minute)
```

```python
df.index[2]
```

```python
df['month']=[m.month for m in df.index]
df['year']=[y.year for y in df.index]
df.head(5)
```

```python
df2=df.pivot(index='year', columns='month',values='Avila Adobe')
df2
```

```python
plt.figure(figsize=(10,3))
sns.heatmap(df2, cmap='Greens')
plt.show()
```

```python
plt.figure(figsize=(10,3))
sns.heatmap(df2, cmap='RdGy_r', linewidths=0.654,linestyle='--',square=True, linecolor='black')

l=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
plt.xticks(ticks=np.arange(12)+.5, labels=l)
plt.title('Muzeyga tashriflar soni')
plt.show()
```
