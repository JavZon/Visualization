```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
np.random.seed(0)
scores = np.random.normal(loc=70, scale=10, size=100)  # Normally distributed scores
df=pd.DataFrame({'Scores':scores})
plt.figure(figsize=(8,4))
sns.histplot(data=df,x='Scores', color='skyblue', binwidth=2.536)
plt.title('Distribution of Exam Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.show()
```

```python
np.random.seed(0)
scores = np.random.normal(loc=70, scale=10, size=100)  # Normally distributed scores
df=pd.DataFrame({'Scores':scores})
```

```python
df['Subject'] = ['Math'] * 34 + ['English'] * 21 + ['Biology'] * 30 + ['Science'] * 15
df
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data=df,x='Scores',hue='Subject', color='skyblue', multiple='stack' , bins=35)  #multiple{“layer”, “dodge”, “stack”, “fill”}
plt.title('Distribution of Exam Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.show()
```

```python
df2=df[(df['Scores']>70.1) & (df['Scores']<77.1)]
plt.figure(figsize=(8,4))
sns.histplot(data=df2,x='Scores',hue='Subject', color='skyblue', multiple='stack')  #multiple{“layer”, “dodge”, “stack”, “fill”}
plt.title('Distribution of Exam Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.show()
```
